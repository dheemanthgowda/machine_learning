```python
data1=data.dropna(how="any",axis=0)
data1.shape
```

```python
data1['classification'] = data1.classification.map({'benign':0, 'malware':1})
```

```python
data1["classification"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.axis("equal")
plt.show()
```

```python
malware1=data.loc[data['classification']=='malware']
malware1["classification"].head()
```

```python
corr=data1.corr()
corr.nlargest(35,'classification')["classification"]
```

```python
x=data1.drop(["hash","classification",'last_interval','min_flt','millisecond','gtime','stime','free_area_cache','total_vm','state'],axis=1)
x.shape
```

```python
y=data1["classification"]
y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=5,p=2)
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
pred
```

```python
model.score(X_test,y_test)
```

```python
result=pd.DataFrame({
    "Actual_Value":y_test,
    "Predict_Value":pred
})
```

```python
result
```

```python

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt     

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['benign', 'malware']); ax.yaxis.set_ticklabels(['benign', 'malware']);
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
results = confusion_matrix(y_test,pred) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(y_test,pred)) 
print ('Report : ')
print (classification_report(y_test,pred)) 
```
